# Plotting antenna parameters

In the ``tools`` sub-package is a module called ``plot_antenna_params`` which can be used to plot parameters from a model containing an antenna. The antenna in the model must be fed using the ``#transmission_line`` command and can optionally, include a receiver antenna with a ``#rx`` command. The module takes an argument which is the name of the output file and several optional arguments:

* ``--tl-num`` is the transmission line number of the transmitting antenna (default=1)
* ``--rx-num`` is the receiver number for the receiver antenna (default=None)
* ``--rx-component`` is the electric field component to analyse for the receiver antenna (default=None)

For example (to use the module outside this notebook) to plot the antenna parameters from the wire dipole antenna:

    python -m tools.plot_antenna_params user_models/antenna_wire_dipole.out

You can use the following code to experiment (in this notebook) with plotting different field/current components and FFTs.

```python
%matplotlib inline
import os

from tools.plot_antenna_params import calculate_antenna_params, mpl_plot

filename = os.path.join(os.pardir, os.pardir, 'user_models', 'antenna_wire_dipole_fs.out')
antennaparams = calculate_antenna_params(filename)
plt = mpl_plot(**antennaparams)
```
